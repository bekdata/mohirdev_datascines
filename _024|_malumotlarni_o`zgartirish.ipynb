```python
df1 = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df1.head()
```

```python
df1['num-of-cylinders'].unique()
```

```python
# har bilan yozilgan sonlarni raqam ko`tinishiga almashtirish
dict = {'four':4, 'six':6, 'five':5, 'three':3, 'twelve':12, 'two':2, 'eight':8}
df1['cylinders'] = df1['num-of-cylinders'].map(dict)
```

```python
# yangilangan dg
df1.drop(['num-of-cylinders'], axis=1)   # [num-of-sylinders] - ustunini olib tashlash
```

```python
# almashgan ustun ma`lumotlarining ko`rinishi
df1[['num-of-cylinders', 'cylinders']]
```
