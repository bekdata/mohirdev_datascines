```python
import numpy as np
import pandas as pd
```

```python
data = {'Yil':[2021, 2020, 2019, 2018, 2017, 2016, 2015, 2010],
       'Aholi soni':[33.9, 33.5, 32.9, 32.5, 31.9, 31.4, 30.9, 28.5],
       'Temp':[1.54, 1.48, 1.56, 1.62, 1.65, 1.66, 1.64, 1.53],
       'Zichlik':[79.77, 78.68, 77.53, 76.34, 75.13, 73.91, 72.71, 67.03]}

data
```

```python
df = pd.DataFrame(data)   # ma`lumotni dataframe ko`rinishiga o`tkazish
df
```

```python
df.index = [21,20,19,18,17,16,15,10]   # index qiymatlarini almashtirish
df
```

```python
df.columns = ['yil', 'axoli', 'temp', 'zichlik']   # ustun qiymatlarini almashitirish
df
```

```python
df.reindex([15,10,21,20,17,16])      # xoxlagan index qiymtlarini chiqarish
```

```python
df.reindex(columns=['axoli', 'zichlik'])    # ustun malumotlarini alohida qilish
```

```python
#locp[[]] - metodidan foydalanish
df.loc[[10,20,21], ['axoli', 'zichlik']]
```

### loc va iloc parametrlaridan foydalanish

```python
df.loc[[2015]]   # 2015 qator qiymatini olish
```

```python
df.loc[[2015,2020]]   # ko`rsatilgan qator malumotlarini chiqarish
```

```python
df.loc[[2020, 2019,2018], ['Uzbekistan', 'Kazakhstan', 'Tajikistan']]   # ustun malumotlari qator bilan birga qo`llaniladi
```

```python
df.iloc[[0,2,4]]    # 0,2,4 qatorlarni chiqarish   qator ma`lumotlarini index qiymati bo`yicha chiqarish
```

```python
df.iloc[[0], [0,1,2,3]]    # birinchi qatorning 1,2,3 va 4 ustun ma`lumotlarini chiqaring
```
