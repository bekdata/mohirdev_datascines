```python
import numpy as np
import pandas as pd
```

```python
# faylni chaqirib olish
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv", index_col=0)
df.head()
```

```python
# data dagi qiymatlarni berish
df['company'].value_counts()
```

```python
df_group = df.groupby('company')
df_group.value_counts()
```

```python
# dataning ikki ustulari o`rtasida solishtirish amalini bajarish
# horsepower ustunining mos qiymatiga price ning mos qiymatini qo`yish
df.groupby('horsepower').price.max() #max narxlarni chiqaramiz
```

```python
# narx ustun bo`yicha o`rta qiymat chiqarish
df.groupby('company').price.mean()
```

```python
# berilgan ustunning mos qiymatlariga mos max qiymatni hisoblash
df.groupby('horsepower').max()
```

```python
# Qo`shimcha bajarish uchun amal kiritish
df.groupby('body-style').price.agg([len, min, max, np.mean])
```

```python
# companiya bo`yicha minimal qiymatlar
df_comp = df.groupby('company')
df_comp.min().head()
```
