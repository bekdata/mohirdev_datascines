```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population_duplicates.csv', usecols=['country','pop2021','area'], index_col='country')
df.head()
```

```python
df.shape
```

```python
# mavjud bir xil bo`lgan qatorlar
sum(df.duplicated())
```

```python
df.drop_duplicates(inplace=True)
df.head()
```

```python
df.shape
```

```python
# index qiymatlarini alifbo tartibida tartiblaydi
df.sort_index(inplace=True)
df.head()
```

```python
# bitta hududning malumotni olish
df.loc['China']
```

```python
# bir nechta qiymat ma`lumotlarni df qiymatda olish
df.loc[['Uzbekistan', 'China']]
```

```python
# malumotlarni ma`lum bir oraliqga qarab joylashtirish
bins = [0,100,1000,5000,10000,15000]

population = df.pop2021
```

```python
# bir yildagi axoli ma`lumotlari
df.pop2021
```

cut() - ma`lumotni berilgan qiymatlar oralig\'ida kesib oladi

```python
oraliq = pd.cut(population, bins)
print(oraliq)
```

```python
oraliq.value_counts()
```

```python
df['popBins'] = oraliq
df
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv", index_col=0)
df.head()
```

```python
# data haqida umumiy ma`lumot
df.describe()
```

```python
# bitta ustun bo`yicha malumot
df.price.describe()
```

```python
bin_price = [0, 10000, 20000, 50000]
# mana shu oraliqni ajratish

ran = pd.cut(df.price, bin_price)
ran.value_counts()
```

```python
# yangi ustunni dataga qo`yish
df['price_bin'] = ran
df.head()
```

```python
# ma`lumotlarni guruhlarga ajratishni dasturni o`ziga qoldiish
data = pd.qcut(df.price, 5)
data.value_counts()

# bo`linishni chiqarish va dataga qo`shissh

df['data'] = data
df
```
