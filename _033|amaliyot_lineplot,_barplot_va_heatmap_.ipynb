```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
data = pd.read_html('https://en.tutiempo.net/climate/ws-384570.html')
df = data[0]
df.sample(10)
```

```python
df['T'] = df['T'].map(lambda x: x.replace('-', '0'))
df['TM'] = df['TM'].map(lambda x: x.replace('-', '0'))
df[['T', 'Tm', 'PP', 'V', 'RA', 'SN', 'TS', 'FG', 'TN', 'GR' ]] = df[['T', 'Tm', 'PP', 'V', 'RA', 'SN', 'TS', 'FG', 'TN', 'GR' ]].map(lambda x: x.replace('-', ' '))
```

```python
df.head()
```

```python
df.fillna(method='ffill', inplace=True)
df.fillna(method='bfill', inplace=True)
```

datalardan olingan grafik qiymatlarini chiqarish


```python
plt.figure(figsize=(16,6))
sns.lineplot(data=df)
plt.title("Toshkent shahri yillik ob-havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.show()
```
