```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/merc.csv")
df.head()
```

```python
df = df[df['transmission'] != "Other"]
```

```python
df = df[df.year > 2010]
sns.scatterplot(data=df, x = 'year', y='price', hue = 'transmission', style = "fuelType")
plt.show()
```

Histplot

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x = 'price')
plt.show()
```

Distplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='price', kde=True)
plt.show()
```

```python
sns.displot(data=df, x='mileage', kde=True, height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```
