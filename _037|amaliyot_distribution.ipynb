```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/uybor_scrapping.csv')
df.head()
```

```python
# data grafigi chiqarish
plt.figure(figsize=(16,6))
sns.histplot(data=df, x='price')
plt.show()
```

```python
df.shape
```

```python
df = df[df.price < 200000]
df.head()
```

```python
df.shape
```

```python
# uylarning maydoni uchun taqsimot grafigi
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='size')
plt.show()
```

```python
# xonalari soni 6 dan katta bo`lgan qiymatlarni tashlab yuborish
df = df[df.rooms < 6]

# xonalar soni bo`yicha taqsimot grafigi
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='rooms')
plt.show()
```

```python
# uy narxining xonalarga bog`liqlik grafigi
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='price', hue='rooms')
plt.show()
```

```python
sns.displot(data=df, x='price', kind='kde', height=5, aspect=2)

#@title Kutilgan natija

sns.displot(data=df, x='size', kind='kde', height=5, aspect=2) # height va aspect grafik hajmini berish uchun
plt.xticks(rotation=90)
plt.show()
```
