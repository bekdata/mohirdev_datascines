```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
df.shape
```

```python
df = df[df['year'] > 2000]

# yil va narx ustunlari o`rtasidagi munosabat
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='year', y='price', hue='transmission', s =100)
plt.show()
```

```python
df = df[df.fuelType != 'Other']  # Ustundan kerak bo`lmagan ma`lumotlarni olib tashlash

plt.figure(figsize=(10, 6))
# bir necha ustunlar o`rtasida bog`liqlik
sns.scatterplot(data=df, x="year", y="mileage",hue = "transmission",style = "fuelType", s = 100)
plt.show()
```

```python
# narx va bosib o`tgan masofalar o`rtasida bog`liqlikni aniqlash
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x=df['mileage'], y=df['price'])
plt.show()
```
