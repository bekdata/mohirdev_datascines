```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x = 'size', y = 'price')
plt.show()
```

```python
# jadvalga kerak bo`lmagan qiymatlarni olib tashlash
df = df[df['size'] < 250]

# qolgan qism uchun funksiya grafigini davom ettirish
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='size', y='price')
plt.show()
```

```python
df = df[df['size'] < 250]

plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='size', y='price')
plt.show()
```

```python
# uy narx, maydon va xonalar soni bo`yicha bog`liqlikni ko`rsatish
df = df[df['size'] < 250]

# qolgan qism uchun funksiya grafigini davom ettirish
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='size', y='price', hue='rooms', style="level", s=100)
plt.show()
```

```python
df = df[df['size'] < 250]

# narx va o`lchamlarning uy joylashgan manzil o`rtasidagi bog`liqligi
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='size', y='price', hue='district')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(data=df, x='district', y='price')
plt.xticks(rotation=90)
plt.show()
```
