### groupby() metodi haqida mavzu bo`ladi

```python
import pandas as pd
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

```python
# isnull() - data mavjud bo`lmagan (NaN) elementlar o`rniga True qiymat chiqaradi
df.isnull()
```

```python
# har bir ustun ma`lumotlarida nechta NaN qiymat borligini hisoblash
df.isnull().sum()
```

```python
# datada mavjud bo`lmagan qiymatlarni % hisobida chiqarish
nulls = df.isnull().sum()
rows = df.shape[0]
(nulls/rows) *100
```

```python
# dropna()
print(df.shape)    #  [oldingi o`lcham]
df2 = df.dropna(axis=0)    # qator xatolik bo`yicha hisoblash
df2.shape    # tozalangan o`lcham
```

```python
df2
```

```python
df2 = df.dropna(axis=1)   # ustun ma`lumotlariga ko`ra filtrlash
df2
```

```python
# dropna() bilan o`chib ketadigan qatorlarga chegara qo`yish
df.dropna(axis=0, thresh = 5).head()
```

```python
# malumoti to`liq bo`lmagan ustunlar o`rtasida dropna qilish
df.dropna(subset=['horsepower','wheel-base']).head()
```
