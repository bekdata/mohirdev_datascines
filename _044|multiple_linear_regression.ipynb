```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/housing_LR.csv', index_col=0)
df.head(10)
```

```python
data = df[df.district == 'Юнусабадский']
data.head()
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(data, test_size=0.2, random_state=34)
```

```python
from sklearn import linear_model
MLR_model = linear_model.LinearRegression()

x_train = np.asanyarray(train_set[['size','max_levels']])
y_train = np.asanyarray(train_set['price'])

MLR_model.fit(x_train,  y_train)

print("Coeficients: ", MLR_model.coef_)
print("theta0: ", MLR_model.intercept_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.pairplot(data)
plt.shov()
```

```python
x_test = np.asanyarray(test_set[['size', 'max_levels']])
y_test = np.asanyarray(test_set['price'])
```

```python
y_predict = MLR_model.predict(x_test)
```

```python
from  sklearn.metrics import mean_absolute_error, mean_squared_error

MAE = mean_absolute_error(y_test, y_predict)
RMSE = np.sqrt(mean_squared_error(y_test, y_predict))

print("MAE: ", MAE)
print("RMSE: ", RMSE)
```
