```python
import pandas as pd
import numpy as np
```

```python
# series elementlar va ular bilan amallar
s1 = pd.Series([1,2,3,4], index = ['a','c','d','e'])
s1
```

```python
s2 = pd.Series([4,3,2,1], index = ['a','b','c','d'])
s2
```

```python
s1 + s2
```

```python
# dataframe va ular bilan birga amallar bajarish
df1 = pd.DataFrame(np.arange(9.).reshape(3,3), columns=list('bcd'), index=['Olma','Anor','Uzum'])
df2 = pd.DataFrame(np.arange(12.).reshape(4,3), columns=list('abc'), index=['Olma','Anor','Qovun','Anjir'])
```

```python
df1 + df2
```

```python
df1-df2
```

```python
df2.add(df1, fill_value = 0)    # qiymat yo`q bolgan joylarni 0 bilan to`ldiradi
```

```python
df = df1+ df2

df
```

```python
df.fill_value = 0
df.mean()   # mar bir ma`lumot turi bo`yicha o`rtacha qiymatni qaytaradi
```

```python
df1.apply(np.sum, axis = 1)
```

```python
dff = lambda x: x.max() - x.min()
df.apply(dff)     # shu malumotlarni orasiga apply orqali funksiyamizni qo`qish
```
