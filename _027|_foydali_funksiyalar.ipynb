```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/MOCK_USER_DATA.csv')
df.head()
```

```python
# data da ma`lumotlar oralig`ini tekshirib olish
df.age.unique()
```

```python
# mantiqan mumkin bo`lmagan qiymatlarn olib tashlaymiz
df[df['age'] > 120]
```

```python
# yosh qiymati hech qachon ham 0 dan kichhik bolmaydi
df[df['age'] < 0]
```

```python
# manfiy bo`lgan qiymatlarni musbat qiymatga o`tkazamiz
df['age'] = df['age'].apply(np.abs)
df.head()
```

```python
# malumotda 0 dan kichik qiymat qolmaganini ko`rishimiz mumkin
df[df['age'] < 0]
```

```python
# datadan shart bo`lmay qolgan qiymatlarni ajratib olish
df_1 = df[df['age'] > 120].index
df_1
```

```python
df.drop(index = df_1, inplace=True)
```

```python
# bundan shartga to`g`ri kelmaydigan qiymatlarning o`chirilib yuborilganini ko`rishimiz mumkin
df[df['age'] > 120]
```

```python
# data dan faqat Female qiymatlari bor qatorlarni olish
df[df['gender'].isin(["Female"])]
```

```python
# data dan faqat male qiymatlarni ajratish
df[df['gender'].isin(['Male'])]
```
