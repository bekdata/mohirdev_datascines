```python
# sinx() funksiya grafigini tasvirlash
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
```

```python
plt.figure(figsize=(8,4))
plt.plot(x, y, linewidth = 2, linestyle = "--", color = 'r')
plt.title('six(x) funksiya grafigi')   # grafikka nom berish
plt.xlabel("x oqning nomi")
plt.ylabel("y o`qnng nomi")
plt.grid()   # grafik ichidagi panjara
plt.show()    # grafik keraksiz qismlarni olib tashlash
```

```python
# ikkita funksiya grafiklaini birlashtirish
import matplotlib.pyplot as plt
import numpy as np

# sin(x) funksiyasi
x = np.linspace(0, 10, 100)
y = np.sin(x)

# sonlar ro'yxati
sonlar = np.arange(0, 1 ,0.1)

plt.figure(figsize=(8, 4))

# sin(x) funksiyasini chizish
plt.plot(x, y, linewidth=2, linestyle="--", color="r", label="sin(x)")

# sonlar ro'yxatini chizish
plt.plot(sonlar, label="sonlar")

# grafikka nom berish
plt.title("sin(x) funksiya grafigi va sonlar")
plt.xlabel("x oqning nomi")
plt.ylabel("y o`qnng nomi")
plt.grid(True)  # grafik ichidagi panjara

# Legend qo'shish
plt.legend(loc=0, fontsize="x-large")

# Grafikni ko'rsatish
plt.show()

```
