```python
import pandas as pd
import numpy as np
```

```python
# darsga kerakli kutubxonalarni chaqirib olish
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0)
df.head()
```

```python
# index qiymatni df ustunidan foydalanish
df.head()
```

```python
# data o`lchamini aniqlash
df.shape
```

```python
# seaborn kutubxonasi yordamida grafikni chiqarish
sns.lineplot(data=df)
```

```python
# grafika qo`shimcha ma`lumotlar qo`shish
plt.figure(figsize=(10,6))
plt.title("Ma`lumot haqida grafik")
sns.lineplot(data=df)
plt.xlabel("Yillar")
plt.xticks(rotation=90)
plt.ylabel("Odamlar soni")
plt.grid()
plt.show()
```

```python
# bitta ustun malumotlarini grafik ko`rinishi
plt.figure(figsize=(10,6))
sns.lineplot(data=df[['Avila Adobe', 'Firehouse Museum']])   # ikkika ustunni ajratib grafik ko`rinishini chiqarish
plt.xticks(rotation=90)
plt.yticks(rotation=90)    # x va y o`qlarda qiymatlarni vertikal va gorizantal ko`rinijshga o`tkazish
plt.grid()
plt.show()
```
